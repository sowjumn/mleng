```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('yelp.xlsx')
```

```python
df = xls.parse('yelp_data')
```

```python
df_city = xls.parse('cities')
```

```python
df_state = xls.parse('states')
```

```python
df = pd.merge(left = df, right = df_state, how = 'inner', left_on = 'state_id', right_on = 'id')
```

```python
df.head()
```

```python
df = pd.merge(left = df, right = df_city, how = 'inner', left_on = 'city_id', right_on = 'id')
```

```python
df.head()
```

```python
del df['id_x']
```

```python
del df['id_y']
```

```python
bars_rest = df['category_0'].isin(['Bars','Restaurants'])
```

```python
df_bars_rest = df[bars_rest]
df_bars_rest.head()
```

```python
pivot_state_cat = pd.pivot_table(df_bars_rest, index=['state','city','category_0'], values=['review_count'], aggfunc=[np.sum])
```
